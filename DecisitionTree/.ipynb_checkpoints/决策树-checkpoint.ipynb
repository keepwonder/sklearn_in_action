## 分类树

```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
wine = load_wine()
```

```python
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.3)
```

```python
clf = DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)
score
```

```python
import graphviz
from sklearn.tree import export_graphviz
```

```python
feature_name = ['酒精','苹果酸','灰','灰的碱性','镁','总酚','类黄酮','非黄烷类酚类','花青素','颜色强度','色调','od280/od315稀释葡萄酒','脯氨酸']
```

```python
import graphviz
from sklearn.tree import export_graphviz
```

```python
dot_data = export_graphviz(clf
                           , out_file=None
                           , feature_names=feature_name
                           , class_names = ['琴酒', '雪莉', '贝尔摩德']
                           , filled=True
                           , rounded=True
                          )
graph = graphviz.Source(dot_data)
graph
```

## 回归树

```python

from sklearn.datasets import load_boston
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
```

```python
boston = load_boston()
regressor = DecisionTreeRegressor(random_state=0)
cross_val_score(regressor, boston.data, boston.target, cv=10, scoring='neg_mean_squared_error')
```

### 一维回归的图像绘制

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 *(0.5 - rng.rand(16))
```

```python
reg1 = DecisionTreeRegressor(max_depth=2)
reg2 = DecisionTreeRegressor(max_depth=5)
reg1.fit(X,y)
reg2.fit(X,y)
```

```python
X_test = np.arange(0.0,  5.0, 0.01)[:, np.newaxis]
y1 = reg1.predict(X_test)
y2 = reg2.predict(X_test)
```

```python
plt.figure()
plt.scatter(X, y, s=20, edgecolors='black', c='darkorange', label='data')
plt.plot(X_test, y1, color='cornflowerblue', label='max_depth=2', linewidth=2)
plt.plot(X_test, y2, color='yellowgreen', label='max_depth=5', linewidth=2)
plt.xlabel('data')
plt.ylabel('target')
plt.legend()
plt.show()
```
